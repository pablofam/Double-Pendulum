```python
from pylab import *
import cv2
```

```python
cap = cv2.VideoCapture("IMG_0359.mov")
```

```python
cap
```

```python
data=[]
while(cap.isOpened()):
    ret, frame = cap.read()
    cv2.imshow('frame',frame)
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    (minVal, maxVal, minLoc, maxLoc) = cv2.minMaxLoc(gray)
    data+=[maxLoc]

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

data is a list of x,y of brightest spot

```python
x,y=list(zip(*data))
plot(x)
plot(y)
figure()
plot(x,y)
```
